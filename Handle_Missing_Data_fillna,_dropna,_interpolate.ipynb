```python
# the day column data type is in string form convt. in date type first
df=pd.read_csv('weather_data.csv',parse_dates=['day'])
```

```python
type(df.day[0])
```

```python
df.set_index('day',inplace=True)
```

```python
df
```

```python
new_df=df.fillna(0)
new_df
```

```python
new_df=df.fillna({
    'temperature':0,
    'windspeed':0,
    'event':'no event'
})
new_df
```

```python
new_df=df.fillna(method='ffill')
new_df
```

```python
new_df=df.fillna(method='bfill')
new_df
```

```python
new_df=df.fillna(method='bfill',axis='columns')
new_df
```

```python
new_df=df.fillna(method='ffill',limit=1)
new_df
```
